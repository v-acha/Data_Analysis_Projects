Getting team names: loop through a predefined list of team names, cities, or abbreviations. Since statsapi.lookup_team() retrieves information based on the lookup value, you need to provide an appropriate lookup value for each team.
1. Create a list of team names, cities, or abbreviations. You can use the team names for simplicity.
2. Use statsapi.lookup_team() within a loop to fetch information for each team.
3. Collect the information into a list or DataFrame for easier handling and display.

```python
# List of MLB team names
teams = [
    "Arizona Diamondbacks", "Atlanta Braves", "Baltimore Orioles", "Boston Red Sox", "Chicago White Sox",
    "Chicago Cubs", "Cincinnati Reds", "Cleveland Guardians", "Colorado Rockies", "Detroit Tigers",
    "Houston Astros", "Kansas City Royals", "Los Angeles Angels", "Los Angeles Dodgers", "Miami Marlins",
    "Milwaukee Brewers", "Minnesota Twins", "New York Yankees", "New York Mets", "Oakland Athletics",
    "Philadelphia Phillies", "Pittsburgh Pirates", "San Diego Padres", "San Francisco Giants",
    "Seattle Mariners", "St. Louis Cardinals", "Tampa Bay Rays", "Texas Rangers", "Toronto Blue Jays",
    "Washington Nationals"
]

# Initialize an empty list to store team information
team_info = []

# Loop through each team name and fetch their information
for team in teams:
    info = statsapi.lookup_team(team)
    if info:
        team_info.extend(info)

# Convert the list of dictionaries to a DataFrame
df2 = pd.DataFrame(team_info)
```

```python
df2['fileCode'] = df2['fileCode'].str.upper()
df2.head()
```

```python
df2.to_csv("teamnames.csv")
```
