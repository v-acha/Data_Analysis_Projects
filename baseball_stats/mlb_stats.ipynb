```python
# Fetch the schedule data for a specific date range
sched = statsapi.schedule(start_date='07/01/2018', end_date='07/31/2018')

# Convert the list of dictionaries to a pandas DataFrame
df = pd.DataFrame(sched)

# Filter out rows with missing data in 'home_score' or 'away_score'
df = df[['home_score', 'away_score']].dropna()

# Melt the DataFrame to have a long format for easier plotting
df_melted = df.melt(var_name='score_type', value_name='score')

# Create Altair chart for home_score and away_score distributions
chart = alt.Chart(df_melted).mark_bar().encode(
    x=alt.X('score:Q', bin=alt.Bin(maxbins=30), title='Score'),
    y=alt.Y('count()', title='Frequency'),
    color='score_type:N',
    tooltip=['score_type', 'score']
).properties(
    width=500,
    height=300,
    title='Distribution of Home and Away Scores'
).interactive()

# Display the chart
chart
```

```python
# Fetch the schedule data for a specific date range
sched = statsapi.schedule(start_date='07/01/2018', end_date='07/31/2018')

# Convert the list of dictionaries to a pandas DataFrame
df = pd.DataFrame(sched)

# Filter out rows with missing data in 'home_score' or 'away_score'
df = df[['game_date', 'home_score', 'away_score']].dropna()

# Convert 'game_date' to datetime for better handling in the plot
df['game_date'] = pd.to_datetime(df['game_date'])

# Melt the DataFrame to have a long format for easier plotting
df_melted = df.melt(id_vars='game_date', var_name='score_type', value_name='score')

# Create Altair chart for home_score and away_score scatter plots with lines
scatter_plot = alt.Chart(df_melted).mark_point(point=True).encode(
    x='game_date:T',
    y='score:Q',
    color='score_type:N',
    tooltip=['game_date:T', 'score_type:N', 'score:Q']
).properties(
    width=800,
    height=400,
    title='Home and Away Scores Over Time'
).interactive()

# Display the chart
scatter_plot
```

```python
# Save the heatmap as an HTML file
output_path = 'home_awaydistr.html'  # Specify the desired output file path
chart.save(output_path)

print(f"Home and away was saved as {output_path}")
```
