```python
import pandas as pd
import altair as alt

#reading file
df = pd.read_csv("/content/players_20.csv")
```

## Scatter top 500 players
1. The plot specifically focuses on visualizing height against wages, beginning at 160 cm to avoid clutter, as the lowest player height is 163cm.
2. The x-axis, representing wage_eur, has been carefully adjusted to maintain clarity, as the primary focus is on height adjustments on the y-axis.
3. The .interactive() method allows users to zoom and pan through the scatter plot


```python
# Sort by wage in descending order and take the top 100
df_sorted = df.sort_values(by='wage_eur', ascending=False).head(500)

# Create the scatter plot
scatter_plot = alt.Chart(df_sorted).mark_circle(size=60).encode(
    x=alt.X('wage_eur:Q', title='Player Wage (EUR)'),  # Quantitative scale for wages
    y=alt.Y('height_cm:Q', scale=alt.Scale(domain=[160, df_sorted['height_cm'].max()]), title='Player Height (cm)'),  # Adjusted y-axis to start at 160 for height
    color=alt.Color('club:N', title='Club'),  # Nominal scale for clubs
    tooltip=['short_name:N', 'club:N', 'wage_eur:Q', 'height_cm:Q', 'nationality:N']  # Tooltips for additional info
).properties(
    width=800,
    height=500,
    title='Top 500 Players by Wage: Height vs. Wage by Club and Nationality'
).interactive()

# Display the plot
scatter_plot
```
